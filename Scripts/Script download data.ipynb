```python
import pandas as pd
import requests
from io import BytesIO #permet de stocker en mémoire
from zipfile import ZipFile
```

```python
url="https://www.insee.fr/fr/statistiques/fichier/5393560/fd_eec20_csv.zip" #enquête 2020 en exemple
```

```python
# Télécharge le fichier ZIP
requete = requests.get(url)
zip_df = ZipFile(BytesIO(requete.content)) #créer un fichier ZIP

# Extraire le fichier CSV du ZIP
with zip_df.open(zip_df.namelist()[0]) as extrait:
    EEC = pd.read_csv(extrait, delimiter=";") # Lire le fichier CSV avec pandas

#test
EEC.head(10)
```
